## Display to level wheel settings and images

```python
import base64
from io import BytesIO
from PIL import Image
import matplotlib.pyplot as plt 
import json

jsonstr = ''
with open('../example_5/layouthierarchy.json','r') as f:
    hier_obj = json.load(f)

              
def display_images(image_arr):    
    png_image0 = base64.decodebytes(bytes(image_arr[0], 'utf-8'))     
    png_image1 = base64.decodebytes(bytes(image_arr[1], 'utf-8'))    
    png_image2 = base64.decodebytes(bytes(image_arr[2], 'utf-8'))

    fig=plt.figure(figsize=(8, 8))
    columns = 2
    rows = 2
    
    pil_0 = Image.open(BytesIO(png_image0))
    fig.add_subplot(221)
    plt.imshow(pil_0)

    pil_1 = Image.open(BytesIO(png_image1))
    fig.add_subplot(222)
    plt.imshow(pil_1)

    pil_2 = Image.open(BytesIO(png_image2))
    fig.add_subplot(223)
    plt.imshow(pil_2)
    
def display_settings(level):
    print(f'name: {level["name"]}')
    print(f'isLeaf: {level["isLeaf"]}')
    print(f'cId: {level["cId"]}')   
    print(f'cName: {level["cName"]}')   
    print(f'displayName: {level["displayName"]}')     
    # print(f'hue: {level["hue"]}')          
    print(f'color: {level["color"]}')
    # print(f'expression: unused')
    # print(f'variation: unused')
    # print(f'markerOrder: unused')
    if level["isLeaf"] != True: 
        print(f'calling disp image {level["isLeaf"] == False}')
        display_images(level['image'])      


display_settings(hier_obj)
```

## Display second level settings and images

```python
def display_all_children(level):
    for child in level["children"]:
        display_settings(child)
        print()

display_all_children(hier_obj)
```

## Display lowest level settings (no images at this level)

```python
for child_L0 in hier_obj["children"]:
    display_all_children(child_L0)
```
